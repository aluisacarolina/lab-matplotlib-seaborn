#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# Sample data
x = np.linspace(0.1, 2, 400)  # Generating x values avoiding x=0 for logarithmic scale
y1 = x ** 2
y2 = np.exp(x)

# Creating the plot with normal scale
plt.figure(figsize=(10, 5))

# Normal scale
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Scale')

# Logarithmic scale
plt.subplot(1, 2, 2)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Logarithmic Scale')
plt.xscale('log')
plt.yscale('log')

# Display the plot
plt.tight_layout()
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# Sample data
x = np.linspace(0.1, 2, 400)  # Generating x values avoiding x=0 for logarithmic scale
y1 = x ** 2
y2 = np.exp(x)

# Creating the plot with normal scale
plt.figure(figsize=(12, 5))

# Normal scale
#Plot 1
plt.subplot(1, 2, 1)
plt.plot(x, y1, label='$y = x^2$')
plt.plot(x, y2, label='$y = e^x$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Scale')
plt.legend()

# Logarithmic scale
#Plot 2
plt.subplot(1, 2, 2)
plt.plot(x, y1, label='$y = x^2$')
plt.plot(x, y2, label='$y = e^x$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Logarithmic Scale')
plt.legend()
plt.xscale('log')
plt.yscale('log')

# Display the plot
plt.show()
```

#### From the Fitbit data, we want to visually understand:

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_steps_by_month = fitbit.groupby(['Months_encoded','Months']).agg({'Steps':'median'}).reset_index()
median_steps_by_month
```

```python
plt.figure(figsize = (12,4))
plt.bar(median_steps_by_month['Months'],median_steps_by_month['Steps'])
plt.xticks(fontsize=10,rotation=45)
plt.yticks(fontsize=10)
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.show()
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
median_steps_by_work_or_weekend = fitbit.groupby('Work_or_Weekend').agg({'Steps':'median'}).reset_index()
median_steps_by_work_or_weekend
```

```python
# Create a dictionary to map values to labels
label_mapping = {0: 'Week', 1: 'Weekend'}
# Replace numerical labels with corresponding text labels
labels = median_steps_by_work_or_weekend.index.map(label_mapping)

plt.pie(median_steps_by_work_or_weekend['Steps'], labels=labels)
plt.show()
```

```python
#Create a Folder called "figures"
import os
folder_name = "figures"
os.makedirs(folder_name, exist_ok=True)

#Save pictures
label_mapping = {0: 'Week', 1: 'Weekend'}
# Replace numerical labels with corresponding text labels
labels = median_steps_by_work_or_weekend.index.map(label_mapping)
plt.pie(median_steps_by_work_or_weekend['Steps'], labels=labels)
plt.savefig(os.path.join(folder_name, 'median_steps_by_work_or_weekend.png'))
plt.show()
plt.figure(figsize = (12,4))
plt.bar(median_steps_by_month['Months'],median_steps_by_month['Steps'])
plt.xticks(fontsize=10,rotation=45)
plt.yticks(fontsize=10)
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.savefig(os.path.join(folder_name, 'median_steps_by_month.png'))
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# Define the column pairs
column_pairs = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

# Create scatter plots for each pair of columns
for x, y in column_pairs:
    plt.figure(figsize = (12,4))
    scatter_plot=plt.scatter(fitbit[x], fitbit[y],c=fitbit['Work_or_Weekend'])
    plt.colorbar(scatter_plot, ticks=[0, 1]) #0 week 1 weekend
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'Scatter plot: {x} vs {y}')
    plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# Define bin sizes
bin_sizes = [10, 20, 50]

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Plot histogram for each bin size
for i, bin_size in enumerate(bin_sizes):
    ax = axes[i]
    ax.hist(titanic['Age'], bins=bin_size)
    ax.set_xlabel('Age')
    ax.set_ylabel('Number of individuals')
    ax.set_title(f'Bin Size: {bin_size}')

plt.tight_layout()
plt.show()
```

#### How does the bin size affect your plot?

A larger bin size (10) provides a smoother, more generalized view but may miss finer details.
A smaller bin size (50) offers more detail and granularity but can result in a noisier plot, potentially overemphasizing minor variations
In my opinion the best one is the first where we get an overview of how age is distributed, with a focus on general trends and major age groups.

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# Set the Seaborn style
sns.set_style('whitegrid')

# Create the plot
plt.figure(figsize=(12, 4))
ax = sns.boxplot(x=titanic['Age'], color='lightcoral', width=0.3, fliersize=5, notch=False)

# Add annotations
age = titanic['Age']
median_age = age.median()
q1_age = age.quantile(0.25)
q3_age = age.quantile(0.75)

# Plot annotations
plt.text(median_age, 1, f'Median: {median_age}', horizontalalignment='left', color='black', fontsize=10)
plt.text(q1_age, 1, f'Q1: {q1_age}', horizontalalignment='left', color='black', fontsize=10)
plt.text(q3_age, 1, f'Q3: {q3_age}', horizontalalignment='left', color='black', fontsize=10)


plt.xlabel('Age')
plt.title('Summary Statistics of Age')

# Hide the legend as it is not needed here
plt.legend().set_visible(False)

# Show the plot
plt.show()
```

I choose boxplot because it provides a summary of the distribution including median, quartiles, and potential outliers

#### What does the last plot tell you about the feature `Age`?

The median line inside the box represents the middle value of the dataset. This indicates that half of the ages are below 28.5 and half are above it, giving a measure of central tendency.
Median is closer to Q1, so the distribution is positively skewed.

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
#Violin plot
plt.figure(figsize=(12,6))
sns.violinplot(x=titanic['Age'],color='lightgreen', saturation= 0.5, inner='quart')
plt.title("Summary Statistics and Disftibution of Age")
plt.xlabel("Age")
plt.xticks(ticks=np.arange(0,titanic['Age'].max()+5,5))
plt.show()
```

#### What additional information does the last plot provide about feature `Age`?

The plot shows a peak around the 27/28, suggesting that most individuals in the dataset are in this age range. Not only can we see the median and quartiles, but we can more clearly identify how the ages are distributed around the median and other points.

#### Plot the correlation matrix using seaborn.

```python
corr_matrix = titanic.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True, fmt=".2f", 
            cbar=True, xticklabels=corr_matrix.columns, yticklabels=corr_matrix.columns,
            cmap="coolwarm", linewidths=0.5)

plt.xticks(rotation=45, ha="right")  
plt.yticks(rotation=0)  
plt.title('Correlation Matrix', pad=20)
plt.tight_layout()
plt.show()
```

```python
corr_matrix
```

#### What are the most correlated features?

The highest correlation is between Fare and PClass.
As Pclass increases (moving from 1st class to 3rd class), the Fare generally decreases. 
This means passengers in higher classes (lower Pclass value) tend to pay more for their tickets.
There is also a moderately significant negative correlation between PClass and Age which suggests that older passengers tend to be found in higher classes, 
and younger passengers tend to be more prevalend in lower classes.
The negative correlation between Pclass and Survived, although not too high, also suggests the expected: the people that survived were mostly from higher classes.
